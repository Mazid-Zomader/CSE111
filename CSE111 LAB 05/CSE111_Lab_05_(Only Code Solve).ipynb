```python
class Farmer  :
    def __init__(self , *args) :
        if len(args) == 0 :
            print("Welcome to your farm!")
        else :
            if type(args[0]) == str :
                print(f'Welcome to your farm, {args[0]}!')
            elif type(args[0]) == int :
                print(f'Welcome to your farm. Your farm ID is {args[0]}!')
        self.crop_list = []
        self.fish_list = []
    def addCrops(self , *args) :
        if len(args) == 0 :
            print("No crop(s) added.")
        else :
            print(f'{len(args)} crop(s) added.')
            for index in range(len(args)) :
                self.crop_list.append(args[index])
    def addFishes(self, *args) :
        if len(args) == 0 :
            print("No fish added.")
        else :
            print(f'{len(args)} fish(s) added.')
            for index2 in range(len(args)) :
                self.fish_list.append(args[index2])
    def showGoods(self) :
        if len(self.crop_list) == 0 :
            print("You don't have any crop(s).")
        else :
            print(f'You have {len(self.crop_list)} crop(s):')
            print(", ".join(self.crop_list))
        if len(self.fish_list) == 0:
            print("You don't have any fish(s).")
        else :
            print(f'You have {len(self.fish_list)} fish(s):')
            print(", ".join(self.fish_list))

f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
class TaxiLagbe :
    def __init__(self , num , place) :
        self.number_plate = num
        self.area = place
        self.consumer_list = []
        self.price_list = []
        self.count = 0
        self.total = 0

    def addPassenger(self , *args) :
        def seperator(a):
            v = '1234567890'
            num = ''
            name = ''
            for index in range(len(a)) :
                if a[index] in v :
                    num += a[index]
                elif a[index] == "_":
                    pass
                else :
                    name += a[index]
            num = int(num)
            return num , name


        for index in range(len(args)) :
            self.count += 1
            if self.count <= 4 :
                a , b = seperator(args[index])
                self.consumer_list.append(b)
                self.price_list.append(a)
                print(f'Dear {b}! Welcome to Taxilagbe')
            else :
                print("Taxi Full! No more passengers can be added.")
                self.count -= 1
    def printDetails(self) :
        print(f"Trip info for Taxi number : {self.number_plate}")
        print(f'This trip only cover {self.area} area')
        print(f'Total passenger: {self.count}')
        print(f'Passenger lists:')
        print(", ".join(self.consumer_list))
        self.total = 0
        for index3 in range(len(self.price_list)) :
            self.total += self.price_list[index3]
        print(f'Total collected fare: {self.total} Taka')



taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100',
'Wood_200','Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115', 'Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
class Sphere :
    def __init__(self , nam , rad = 1,col = "White") :
        self.name = nam
        self.radius = rad
        self.volume = (4/3) * 3.1416 * (self.radius**3)
        self.colour = col
    def printDetails(self) :
        print(f'Sphere ID : {self.name}')
        print(f'Colour : {self.colour}')
        print(f'Volume: {self.volume}')
    def merge_sphere(self,*args) :
        merged = 0.00
        for index in range(len(args)) :
            merged += args[index].volume
        self.volume += merged
        for index2 in range(len(args)) :
            if self.colour == args[index2].colour :
                pass
            else :
                self.colour = "Mixed Colour"
                break
        print('Sphere are being merged')


sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```

```python
class Department:
    def __init__(self , dept = 'ChE Department' , sec = 5) :
        self.department = dept
        self.section = sec
        self.average = 0
        self.mega = 0
        print(f'The {self.department} has {self.section} sections.')
    def add_students(self , *args) :
        if len(args) == self.section :
            average = 0
            for index in args :
                average += index
            average = average / self.section
            print(f'The {self.department} has an average of {round(average , 2)} students in each section.')
            self.average = average
            self.mega = average
        else :
            print(f"The {self.department} doesn't have {len(args)} sections.")
    def merge_Department(self , *args) :
        self.mega = self.mega * self.section
        for index in (args) :
            print(f'The {index.department} is merged to {self.department}')
            self.mega += index.average * index.section
        self.mega /= self.section
        return(f'Now the {self.department} has an average of {self.mega} students in each section.')
        
        
# Driver Code
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22,
32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Shopidify :
    def __init__(self , name = 'Guest') :
        self.name = name
        if self.name == 'Guest' :
            print('Welcome to Shopidify')
        else :
            print(f'Welcome {self.name} to Shopidify' )
        self.dict = {}
    def add_to_cart(self , *args) :
        if type(args[0]) == list :
            li =args[0]
            for index in range(0 , len(li) ,2 ) :
                self.dict[li[index]] = li[index + 1]
        else :
            if len(args) == 1 :
                if args[0] in self.dict.keys() :
                    self.dict[args[0]] += 1 
                else :
                    self.dict[args[0]] = 1
            else :
                self.dict[args[0]] = args[1]
    def display_cart(self) :
        print(f'Items in the cart for {self.name} :')
        for key , value in self.dict.items() :
            print(f'- {key}: {value}x')
    def checkout(self) :
        print(f'Checkout completed for {self.name}')
    def display_history(self) :
        print(f'Purchase history for {self.name}')
        print('Transaction 1:')
        for key , value in self.dict.items() :
            print(f'- {key}: {value}x')
        
            
#Test the Shopidify class
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```
