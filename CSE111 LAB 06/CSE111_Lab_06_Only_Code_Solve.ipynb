```python
class Student :
    def __init__(self , nam , id , cg) :
        self.nam = nam
        self.id = id
        self.cgpa = cg
    def setId(self, newID) :
        self.id = newID

class Department :
    def __init__(self , dept) :
        self.department = dept
        self.loc = []
        self.store_id = []
    def findStudent(self , s_id) :
        id = s_id
        flag = False
        for index in range(len(self.loc)) :
            if self.loc[index].id != id :
                pass
            else :
                flag = True
                name = self.loc[index].nam
                id = self.loc[index].id
                cgpa = self.loc[index].cgpa
                break
        if flag:
            print(f"Student Info:")
            print(f"Student Name: {name}")
            print(f"ID: {id}")
            print(f"CGPA: {cgpa}")
        else :
            print("Student with this ID doesn't exist, Please give a valid ID")
    def addStudent(self , *args) :
        for index in range(len(args)) :
            if args[index].id in self.store_id :
                print("Student with the same ID already exists, Please try with another ID")
            else :
                self.loc.append(args[index])
                self.store_id.append(args[index].id)
                print(f"Welcome to {self.department} department, {args[index].nam}")
    def details(self) :
        print(f"Department Name: {self.department}")
        print(f"Number of student: {len(self.loc)}")
        print("Details of the students:")
        for index in self.loc :
            print(f"Student name: {index.nam}, ID: {index.id},cgpa: {index.cgpa}")


s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```
